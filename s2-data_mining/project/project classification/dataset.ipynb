```python
from sklearn.datasets import load_breast_cancer
import matplotlib.pyplot as plt
import numpy as np

# Charger le dataset
data = load_breast_cancer()
y = data.target
class_names = data.target_names

# Compter les occurrences par classe
unique, counts = np.unique(y, return_counts=True)

# Tracer un graphique en barres
plt.figure(figsize=(6,4))
plt.bar(class_names, counts, color=['red', 'green'])
plt.title("Répartition des classes dans le dataset Breast Cancer")
plt.xlabel("Classe")
plt.ylabel("Nombre d'échantillons")
for i, count in enumerate(counts):
    plt.text(i, count + 5, str(count), ha='center')
plt.show()
```

```python
from sklearn.datasets import load_breast_cancer
from imblearn.over_sampling import ADASYN
from collections import Counter

# Charger les données
data = load_breast_cancer()
X = data.data
y = data.target

print("Avant ADASYN :", Counter(y))

# Appliquer ADASYN
adasyn = ADASYN(random_state=42)
X_res, y_res = adasyn.fit_resample(X, y)

print("Après ADASYN :", Counter(y_res))

```

```python
import matplotlib.pyplot as plt
from collections import Counter

# Avant ADASYN
counter_before = Counter(y)
# Après ADASYN
counter_after = Counter(y_res)

# Préparation des données pour le graphe
labels = ['Classe 0', 'Classe 1']
counts_before = [counter_before[0], counter_before[1]]
counts_after = [counter_after[0], counter_after[1]]

x = range(len(labels))

plt.figure(figsize=(8,5))
plt.bar(x, counts_before, width=0.4, label='Avant ADASYN', align='center')
plt.bar([i + 0.4 for i in x], counts_after, width=0.4, label='Après ADASYN', align='center')
plt.xticks([i + 0.2 for i in x], labels)
plt.ylabel('Nombre d\'échantillons')
plt.title('Distribution des classes avant et après ADASYN')
plt.legend()
plt.show()
```
