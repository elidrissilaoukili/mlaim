```python
import numpy as np
from scipy.stats import norm
import pandas as pd
import matplotlib.pyplot as plt
df= pd.read_csv("dataset_taille.csv")
print(df.head())
```

```python
df1 = df["Height"]
print(df1.head())
```

```python
num_bins = 100
# Histograme(PMF) des Tailles
counts, bins = np.histogram(df1, np.linspace(min(df1), max(df1), num_bins))
plt.hist(df1, bins, density=True)
```

```python
num_bins = 100
# Histograme(PMF) des Tailles
counts, bins = np.histogram(df1 ,np.linspace(min(df1), max(df1), num_bins))
plt.hist(df1, bins, density=True)
		
#PDF of Normal Distribution
mean, std = np.mean(df1), np.std(df1)
rvs = np.linspace(mean - 3*std, mean + 3*std, 100)
pdf = norm.pdf(rvs, mean, std)
plt.plot(rvs, pdf, c="g") 
plt.xlabel("Taille(Inches)")
plt.ylabel("Proportion de la taille se situant dans une certaine fourchette")
plt.show()
```

```python
num_bins = 100

# Histograme(PMF) des Tailles
counts, bins = np.histogram(df1 ,
							np.linspace(min(df1),
										max(df1),
										num_bins)
							)
pmf = counts/sum(counts)
cdf = np.cumsum(pmf)
plt.scatter(bins[:-1], cdf, c='b', label="Distribution de la taille CDF")
# CDF of Normal Distribution
mean, std = np.mean(df1), np.std(df1)
rvs = np.linspace(mean - 3*std, mean + 3*std, 100)
cdf = norm.cdf(rvs, mean, std)
plt.plot(rvs, cdf, c='g', label="Distribution normale CDF")
plt.legend()
```
