```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.arima_process import ArmaProcess
```

```python
# Question 1: Simulation du processus AR(1) avec a = 0.6
ar1 = np.array([1, -0.6])
ma = np.array([1])
ar1_process = ArmaProcess(ar1, ma)
X_ar1 = ar1_process.generate_sample(nsample=1000)

plt.figure(figsize=(12, 6))
plt.plot(X_ar1)
plt.xlabel('Temps (t)')
plt.ylabel('Valeurs')
plt.title('Simulation du processus AR(1) avec a=0.6')
plt.grid()
plt.show()
```

```python
# Tracé de l'ACF du processus AR(1)
sm.graphics.tsa.plot_acf(X_ar1, lags=40)
plt.title("ACF du processus AR(1) avec a=0.6")
plt.show()
```

```python
# Question 2: Équivalence avec un modèle MA(∞)
# On génère les coefficients MA jusqu'à un certain ordre
ma_coeffs = [0.6**i for i in range(1, 10)]
print("Coefficients MA(∞) approximés:", ma_coeffs)
```

```python
# Question 3: Simulation du processus AR(2) avec a1=0 et a2=-0.9
ar2 = np.array([1, 0, 0.9])
ma = np.array([1])
ar2_process = ArmaProcess(ar2, ma)
X_ar2 = ar2_process.generate_sample(nsample=1000)

plt.figure(figsize=(12, 6))
plt.plot(X_ar2)
plt.xlabel('Temps (t)')
plt.ylabel('Valeurs')
plt.title('Simulation du processus AR(2) avec a1=0, a2=-0.9')
plt.grid()
plt.show()
```

```python
# Tracé de l'ACF du processus AR(2)
sm.graphics.tsa.plot_acf(X_ar2, lags=40)
plt.title("ACF du processus AR(2) avec a1=0, a2=-0.9")
plt.show()
```
