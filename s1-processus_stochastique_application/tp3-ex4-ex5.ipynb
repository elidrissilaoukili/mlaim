```python
import numpy as np
import matplotlib.pyplot as plt
N=1000; landa=100
X=np.random.exponential(1/landa,N)
Y=3*X;
plt.figure(1);
plt.scatter(X,Y);
Z1=np.corrcoef(X,Y);
```

```python
N=1000; a=0; b=2*np.pi
teta=np.random.uniform(a,b,N) ;
X=np.sin(teta)
Y=np.cos(teta)
plt.figure(4);
plt.scatter(X,Y);
Z2=np.corrcoef(X,Y);
```
