```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Définition de l'intervalle de temps
t = np.arange(1, 201)

# Génération des séries temporelles
T = 100 + 3 * t
S1 = 50 * np.cos(2 * np.pi * t / 3)
S2 = 50 * np.sin((np.pi / 4) + (np.pi * t / 2))
S = S1 + S2

# Perturbation i.i.d. suivant une loi normale
epsilon = np.random.normal(0, 15, size=len(t))

# Série X
X = T + S + epsilon

# Création d'un DataFrame Pandas
df = pd.DataFrame({'t': t, 'X': X})

# Application de la moyenne mobile M3
df['X_M3'] = df['X'].rolling(window=3).mean()

# Représentation graphique
plt.figure(figsize=(12, 6))
plt.plot(df['t'], df['X'], label='Série X', alpha=0.7)
plt.plot(df['t'], df['X_M3'], label='Moyenne mobile M3', linewidth=2)
plt.xlabel('Temps (t)')
plt.ylabel('Valeurs')
plt.title('Série Temporelle et Moyenne Mobile')
plt.legend()
plt.grid()
plt.show()
```
